```python
def suma(operando1, operando2):
  valor3 = operando1 + operando2
  return valor3

def resta(operando1, operando2):
  valor3 = operando2 - operando1
  return valor3
```

```python
class Calculadora():

  def __init__(self, arg1,arg2):
    self.operando1 = arg1
    self.operando2 = arg2

  def sumaEnClase(self):
    valor3 = self.operando1 + self.operando2
    return valor3

  def restaEnClase(self):
    valor3 = self.operando2 - self.operando1
    return valor3
  
  def mult(self):
    valor3 = self.operando1 * self.operando2
    return valor3
```

```python
Calculadora(1, 2).restaEnClase()
```
